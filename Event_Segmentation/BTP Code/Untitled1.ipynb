```python
har_sen=list()
for i in range(1, 18):
    file = open(r'C:\Users\Sourav\BTP Code\Harry_Potter\v1c'+str(i)+'grp', 'rb')
    data = pickle.load(file)
    if 'Harry' in data:
        har_sen+=data['Harry']
    file.close()
```
